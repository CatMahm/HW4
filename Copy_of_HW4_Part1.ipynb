```python
# Import required libraries
import pandas as pd
import plotly.express as px

# Read the olympic_medals.csv file from your Google Drive path
olympic_medals = pd.read_csv('/content/drive/MyDrive/DATA/olympic_medals.csv')

# Filter data for years since 2000
olympic_medals_filtered = olympic_medals[olympic_medals['Year'] >= 2000]

# Create the parallel categories plot with a nice color scale
fig = px.parallel_categories(
    olympic_medals_filtered,
    dimensions=['Gender', 'Medal'],
    color='Year',  # Use Year column to color the lines
    title='Proportions of Medal Types by Gender Since 2000 (Colored by Year)',
    labels={'Gender': 'Gender', 'Medal': 'Medal Type', 'Year': 'Year'},
    color_continuous_scale=px.colors.sequential.Viridis  # Nice, vibrant color scale
)

# Adjust layout for better visibility
fig.update_layout(
    width=800,  # Slightly larger figure
    height=600,
    title_x=0.5  # Center the title
)

# Display the plot
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read CSV
df = pd.read_csv('/content/drive/MyDrive/DATA/2016elections.csv')

# Filter for selected states
states = ['AR', 'MI', 'CA', 'WI']
df_filtered = df[df['st'].isin(states)]

# Prepare plot data
df_plot = df_filtered[['st', 'pct_trump', 'pct_clinton', 'pct_johnson', 'pct_other']].set_index('st')

# Use a custom color palette
colors = ['#e74c3c', '#3498db', '#f1c40f', '#95a5a6']  # Trump (red), Clinton (blue), Johnson (yellow), Others (gray)

# Use seaborn styling
plt.style.use('Solarize_Light2')

# Create the stacked bar chart
ax = df_plot.plot(kind='bar', stacked=True, color=colors, figsize=(10, 6), edgecolor='black')

# Add value labels to each segment
for i, state in enumerate(df_plot.index):
    cumulative = 0
    for j, candidate in enumerate(df_plot.columns):
        value = df_plot.loc[state, candidate]
        if value > 2:  # Only label if > 2% to avoid clutter
            ax.text(i, cumulative + value / 2, f"{value:.1f}%", ha='center', va='center', fontsize=8, color='black')
        cumulative += value

# Customize plot
plt.title('2016 Presidential Election Vote Percentages by State', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('State', fontsize=12)
plt.ylabel('Vote Percentage (%)', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(['Trump', 'Clinton', 'Johnson', 'Others'], title='Candidate', fontsize=9, title_fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show plot
plt.show()
```
